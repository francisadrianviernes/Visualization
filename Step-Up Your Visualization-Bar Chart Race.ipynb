```python
import bar_chart_race as bcr
import pandas as pd
import numpy as np

#Supress Warning
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
df = pd.read_csv('data/GDP_2010_PPP.csv', encoding='utf-8-sig', skipfooter=52)
df.head()
```

```python
years = df.columns.to_list()[4:]
preprocessed = df.melt(id_vars=['Country Name'], value_vars=years, var_name='year', value_name='gdp_ppp')
```

```python
preprocessed['year'] = \
preprocessed['year'].str.replace(r'\[.*\]','',regex=True).str.strip()
preprocessed.head()
```

```python
df = preprocessed.pivot(index='year',
                  columns='Country Name',
                  values='gdp_ppp')
df.head()
```

```python
cols = df.columns.to_list()
for i in cols:
    df[i] = pd.to_numeric(df[i], errors='coerce')
    df[i] = df[i].div(1000000000, axis=0)
```

```python
bcr.bar_chart_race(df, 
                   n_bars=10,
                  title = "GDP in Billions USD (2010 PPP)",
                  cmap='dark2',
                  filter_column_colors=True) #This ensures that the colors in your colormap gets reserved for rows that do make it into the chart. Not all of the rows will make it to the top 10 so ensure that this is set to True.
```
