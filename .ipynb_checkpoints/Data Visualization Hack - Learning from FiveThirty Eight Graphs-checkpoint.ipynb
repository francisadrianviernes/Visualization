FiveThirtyEight (sometimes written as 538), is a website that produces opinion polls on topics such as politics, economics, and sports. It takes its name from the number of electors in the United States electoral college. But one of the things that contribute to their popularity is the effectiveness of their visualization in relaying poll results.

We can learn a thing or two in improving their visualization to make our presentations more professional-looking and captivating. We will refer to the graphs created by FiveThirtyEight as FTE graphs for the rest of the article.

## PRELIMINARIES

```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.style as style

%matplotlib inline
```

#  IMPORTING DATASET

```python
df = pd.read_csv('data\movies.csv', encoding='latin-1')
df.head()
```

Assume for example that we want to examine how the budget of the different genres for movies across the years. For this, let's use the dataset from Kaggle: 

https://www.kaggle.com/rounakbanik/the-movies-dataset

Examining the budget across the years might reveal the changing preferences and complexity of producing a movie in a particular genre.

```python
# Some data preprocessing to make the manipulation easiers
trial = df.groupby(['year', 'genre'])['budget'].sum().reset_index()
trial.head()

trial = trial.pivot(index="year", columns="genre").dropna(axis = 1, how="any").reset_index()
# trial.iloc[0,].index.get_level_values('genre')
trial.iloc[0,1:]
```

```python
initial_graph = trial.plot(x ='year', y = 'budget')
```

```python
style.available
```

```python
#To start, using the set style allows us to easily elevate the level of our visualization
style.use('fivethirtyeight')
```

Using the packaged style from matplotlib instantly provides certain improvements from the base graph of matplotlib. Some of these are as follows:

  *  background color
  *  grid lines
  *  font
  *  line thickness
  *  removed top and right spines

```python
fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8))
```

We can examine an ideal FTE graph through an example: https://fivethirtyeight.com/features/our-47-weirdest-charts-from-2015/

To bring us closer to the style of an FTE graph, we still need the following components:

  *  **Title**
  *  **Subtitle**
  *  **Signature Bar**
  
Other minor adjustments for the graph, may include the following:

  *  font size for tick labels
  *  number format for the major tick-labels of the y-axis
  *  placing a **bold** horizontal line at y=0
  *  removing x-axis label
  *  legends are replaced by text identifying the graph inline

## STORING THE GRAPH INSIDE A VARIABLE

Storing the graph inside a variable makes it easier for us to access their attributes and to apply some functions on the object.

```python
fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8))

#Accessing the tick parameters and modifying the size
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)


# Customizing the tick labels of the y-axis
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);
```

## BOLDING HORIZONTAL LINE Y=0

```python
fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8))


fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)


# Customizing the tick labels of the y-axis
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);

# Generate a bolded horizontal line at y = 0
fte_graph.axhline(y = 0, color = 'black', linewidth = 2, alpha = .7)
```

## ADD AN EXTRA VERTICAL LINE

```python
fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8))
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);
fte_graph.axhline(y = 0, color = 'black', linewidth = 2, alpha = .7)


# Add an extra vertical line by tweaking the range of the x-axis, since the major ticks of the x-axis is 5, setting xlimits to 1984 and 2016, provides some much needed space in our graph, particularly so that our tick labels do not seem too compressed
fte_graph.set_xlim(left = 1984, right = 2016)
```

## SIGNATURE BAR

The signature bar has the following characteristics:

  * Positioned at the bottom bar
  * The signature bar has a dark grey background
  * Author's name is written in the bar
  * Source of data is included below
  * The text color is matching the same color of the 'fivethirtyeight' style

Adding a signature bar requires a bit of effort on our part. However, once we're done, it will set our visuals apart from the rest and would certainly make it look more professional.

The background parameter for the text method of the graph will accomplish highlighting the entire text. This is what we need to produce the signature "bar". However, this background will only highlight the length of the text. Thus, to produce the alignment that we want, this must mean that we will utilize whitespaces in the text.


So to summarize the important steps for the signature bar, we need the following:
  *  text method with the whitespaces, done through trial and error, to align the signature
  *  the placement of the signature bar, ideally below your lowest x-value, done through trial and error as well

```python
fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8))
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);
fte_graph.axhline(y = 0, color = 'black', linewidth = 2, alpha = .7)
fte_graph.set_xlim(left = 1984, right = 2016)

#Remove the label of the x-axis
fte_graph.xaxis.label.set_visible(False)

# The signature bar
fte_graph.text(x = 1979.8, y = -1000000000,
    s = ' Francis Adrian Viernes                                                          Source: Data Collected by TMDB and GroupLens Accessed via Kaggle ',
               fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey');
```

## TITLE AND SUBTITLE PLACEMENT

As with what we have done with the placement of signature bar, the title and subtitle components are 'text' elements. The placement will have to be done through trial and error of the x- and y- coordinates of the graph. The title and subtitle are found on top of the graph so ideally, the y-coordinates of these texts exceed that of the highest y-value in your dataset.

Because we are concerned about the alignment of our x-values, let's use the one we have for the Signature Bar.

```python
#Code to find the maximum y-value
trial.to_numpy().max()
```

Since the maximum value of our y is around the $4.79B, we can try to check if placing it at 5.2 Billion looks good. 

```python
fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8))
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);
fte_graph.axhline(y = 0, color = 'black', linewidth = 2, alpha = .7)
fte_graph.set_xlim(left = 1984, right = 2016)
fte_graph.xaxis.label.set_visible(False)
fte_graph.text(x = 1979.8, y = -1000000000,
    s = ' Francis Adrian Viernes                                                          Source: Data Collected by TMDB and GroupLens Accessed via Kaggle ',
               fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey');
#Title and Subtitle Placement
fte_graph.text(x = 1980, y = 5200000000, s = "Movie Budget Across Genres Through the Years",
               fontsize = 26, weight = 'bold', alpha = .75);
fte_graph.text(x = 1980, y = 4900000000,
               s = 'A Peak Into Which Movie Genre Has Become More Expensive To Produce Within The Last 30 Years',
              fontsize = 18, alpha = .85);
```

## COLOR SCHEMES

Borrowing from the ideas of the base article we are following, and from this source: https://www.nature.com/articles/nmeth.1618#ref1, choosing colors that are suitable for color-blind readers not only enhances accessibility but is also a good graphic-design practice. A code for the RGB has been prepared by our original article source and luckily for us, it has the same number as our genre above

```python
# Colorblind-friendly colors
colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],
          [240/255,228/255,66/255], [0,114/255,178/255], [213/255,94/255,0], [204/255,121/255,167/255]]

fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8), 
           color= colors) #instead of choosing a colormap, we handpick our colors
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);
fte_graph.axhline(y = 0, color = 'black', linewidth = 2, alpha = .7)
fte_graph.set_xlim(left = 1984, right = 2016)
fte_graph.xaxis.label.set_visible(False)
fte_graph.text(x = 1979.8, y = -1000000000,
    s = ' Francis Adrian Viernes                                                          Source: Data Collected by TMDB and GroupLens Accessed via Kaggle ',
               fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey');

#Title and Subtitle Placement
fte_graph.text(x = 1980, y = 5200000000, s = "Movie Budget Across Genres Through the Years",
               fontsize = 26, weight = 'bold', alpha = .75);
fte_graph.text(x = 1980, y = 4900000000,
               s = 'A Peak Into Which Movie Genre Has Become More Expensive To Produce Within The Last 30 Years',
              fontsize = 18, alpha = .85);
```

## REMOVING THE LEGEND AND IDENTIFYING LINES

As with the signature bar, title and subtitle, this portion involves the trial and error for the placement of the text. Instead of a legend, it is now common and an accepted way of saving real estate in graphs, to place texts that simply identify lines or bars in charts. 

To begin, let's remove the legend by the incorporating it into the main code.

```python
# Colorblind-friendly colors
colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],
          [240/255,228/255,66/255], [0,114/255,178/255], [213/255,94/255,0], [204/255,121/255,167/255]]

fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8), 
           color= colors,
           legend= False) #instead of choosing a colormap, we handpick our colors


fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);
fte_graph.axhline(y = 0, color = 'black', linewidth = 2, alpha = .7)
fte_graph.set_xlim(left = 1984, right = 2016)
fte_graph.xaxis.label.set_visible(False)


#Signature Bar
fte_graph.text(x = 1979.8, y = -1000000000,
    s = ' Francis Adrian Viernes                                                          Source: Data Collected by TMDB and GroupLens Accessed via Kaggle ',
               fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey');

#Title and Subtitle Placement
fte_graph.text(x = 1980, y = 5200000000, s = "Movie Budget Across Genres Through the Years",
               fontsize = 26, weight = 'bold', alpha = .75);
fte_graph.text(x = 1980, y = 4900000000,
               s = 'A Peak Into Which Movie Genre Has Become More Expensive To Produce Within The Last 30 Years',
              fontsize = 18, alpha = .85);
```

Next, we use the color-schemes we coded and try to place the text near the lines that they are supposed to identify. To add the colored-labels, we begin the one-by-one placement. There should be an automatic way to do this but an FTE graph is done quite manually to ensure a more effective visuals.  


```python
# Colorblind-friendly colors
colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],
          [240/255,228/255,66/255], [0,114/255,178/255], [213/255,94/255,0], [204/255,121/255,167/255]]

fte_graph = trial.plot(x ='year', 
           y = 'budget', 
           figsize=(12,8), 
           color= colors,
           legend= False) #instead of choosing a colormap, we handpick our colors


fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
fte_graph.set_yticklabels(labels = [-1000000000, 0,  '$1 Billion  ', '$2 Billion  ', '$3 Billion  ', '$4 Billion  ']);
fte_graph.axhline(y = 0, color = 'black', linewidth = 2, alpha = .7)
fte_graph.set_xlim(left = 1984, right = 2016)
fte_graph.xaxis.label.set_visible(False)


#Signature Bar
fte_graph.text(x = 1979.8, y = -1000000000,
    s = ' Francis Adrian Viernes                                                          Source: Data Collected by TMDB and GroupLens Accessed via Kaggle ',
               fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey');

#Title and Subtitle Placement
fte_graph.text(x = 1980, y = 5200000000, s = "Movie Budget Across Genres Through the Years",
               fontsize = 26, weight = 'bold', alpha = .75);
fte_graph.text(x = 1980, y = 4900000000,
               s = 'A Peak Into Which Movie Genre Has Become More Expensive To Produce Within The Last 30 Years',
              fontsize = 18, alpha = .85);

#Colored Labels Indentifying the Line
fte_graph.text(x = 2010, y = 3500000000, s = 'Action', color = colors[0], weight = 'bold',fontsize=16, rotation = 42,
              backgroundcolor = '#f0f0f0');

fte_graph.text(x = 2015, y = 600000000, s = 'Adventure', color = colors[1], weight = 'bold',fontsize=16, rotation = 52,
              backgroundcolor = '#f0f0f0');

fte_graph.text(x = 2014, y = 1200000000, s = 'Animation', color = colors[2], weight = 'bold',fontsize=16, rotation = 52,
              backgroundcolor = '#f0f0f0');

fte_graph.text(x = 2016, y = 500000000, s = 'Biography', color = colors[3], weight = 'bold',fontsize=16, rotation = 52,
              backgroundcolor = '#f0f0f0');

fte_graph.text(x = 2004, y = 1400000000, s = 'Comedy', color = colors[4], weight = 'bold',fontsize=16, rotation = -58,
              backgroundcolor = '#f0f0f0');

fte_graph.text(x = 2016, y = 100000000, s = 'Crime', color = colors[5], weight = 'bold',fontsize=16, rotation = 52,
              backgroundcolor = '#f0f0f0');

fte_graph.text(x = 2002.8, y = 800000000, s = 'Drama', color = colors[6], weight = 'bold',fontsize=16, rotation = 65,
              backgroundcolor = '#f0f0f0');

fte_graph.text(x = 2016, y = -200000000, s = 'Horror', color = colors[7], weight = 'bold',fontsize=16, rotation = 52,
              backgroundcolor = '#f0f0f0');

```

Because the graph we have above have lines that are closely intertwined with one another, an option we used it so place the colored labels on the end of the lines. Doing this, however, extended the graph size and we therefore need to add more whitespaces to our signature bar to align it with the graph.
